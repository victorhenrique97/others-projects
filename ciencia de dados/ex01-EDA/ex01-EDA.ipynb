Resolução para dataset "Breast Cancer"

1) Para esse dataset formulamos questões mais relacionadas à se algum fator ou atributo do tumor tinha como consequência, na sua maioria, se o câncer era maligno ou benigno, além da relação entre dois atributos independentes do target (maligno e benigno).
    - Há relação entre a simetria do tumor e o target?
    - Há relação entre o número de "concave points" e o seu peso?
    - Há relação entre a área do tumor e o target?
    
    
2) Os atributos escolhidos foram os que devem ser utilizados para responder as hipóteses levantadas além de outros atributos que, à primeira vista, aparentam ser úteis para o conjunto de dados no seu contexto da gravidade do câncer. Atributos separados no In[4]
    - simetria: mais próximo de zero mais simétrico.
	- concave points: número de pontos concavos.
	- concavity: peso do atributo de pontos concavos.
	- textura: desvio padrão em uma escala.
	- área.

3) Para todos os atributos utilizamos média e mediana. A moda não faz sentido utilizar pois os atributos são contínuos e raramente se repetem.
   Nesse dataset, há 3 tipos de medidas para cada atributo: "mean", "standard error" and "worst". Utilizamos os campos de "mean" e, apesar de usarmos a média desse atributo, acreditamos que a mediana seja mais adequada já que a média pode ser sensível a pontos extremos.
    Dos blocos In[6] até In[9] realizam a média e a mediana desses atributos.
    

4) De acordo com os resultados, podemos ver que todas as hipóteses podem ser respondidas já que os dados estão relacionados e os valores estão dependentes entre sí, sendo que um influencia no outro.

    - Sim, menor a simetria mais casos malignos
    - Sim, um número maior de concave points estão, geralmente, ligados a um peso grande (e também a câncer maligno)
    - Sim, quanto maior a área, mais casos de malignos.

5) Essa base de dados está estruturada, portanto os dados estão organizados de maneira que é facilmente utilizado para realizar medidas estatísticas e inferir algo.


```python
import numpy as np
import pandas as pd
from sklearn import datasets

cancer = datasets.load_breast_cancer()
cols = cancer['feature_names'].tolist()
cols.append('target')
cancer = pd.DataFrame(data=np.c_[cancer['data'], cancer['target']],
                       columns = cols
                     )

cancer
```

```python
#seprando os atributos que serão utilizados
cancer_atributes = cancer[['mean symmetry','mean concave points', 'mean concavity', 'mean texture', 'mean area', 'target']]
cancer_atributes
```

```python
#separando os datasets
cancer_m = cancer_atributes[cancer_atributes.target == 0.0]
cancer_b = cancer_atributes[cancer_atributes.target == 1.0]
```

```python
cancer_m.mean()
```

```python
cancer_b.mean()
```

```python
cancer_m.median()
```

```python
cancer_b.median()
```
